```python
filtered_df = df[["ID","Name","Overall","Work Rate","Age in 2019","Age in 2023","2019 Value","2020 Value","2021 Value","2022 Value", "2023 Value"]]
filtered_df.head(22)
```

```python
grouped = filtered_df.pivot_table(index="Work Rate", values = ["2019 Value","2020 Value","2021 Value","2022 Value","2023 Value"], aggfunc="mean")
grouped
```

```python


import matplotlib.pyplot as plt

value_cols = ['2019 Value', '2020 Value', '2021 Value', '2022 Value', '2023 Value']
df_long = df.melt(id_vars=['Work Rate'], value_vars=value_cols, var_name='Year', value_name='Value')
df_long['Year'] = df_long['Year'].str.replace(' Value', '', regex=False)

plt.figure(figsize=(14, 7))

wr_colors = {
    'High/ High': '#E65100',    # Strong Orange
    'High/ Medium': '#FF9800',  # Average Orange
    'High/ Low': '#FFCC80',     # Weak Orange
    'Medium/ High': '#1B5E20',  # Strong Green
    'Medium/ Medium': '#4CAF50',# Average Green
    'Medium/ Low': '#A5D6A7',   # Weak Green
    'Low/ High': '#0D47A1',     # Strong Blue
    'Low/ Medium': '#2196F3',   # Average Blue
    'Low/ Low': '#BBDEFB'       # Weak Blue
}

unique_wr = df_long['Work Rate'].unique()

for wr in unique_wr:
    subset = df_long[df_long['Work Rate'] == wr]
    series = subset.groupby('Year')['Value'].mean()
    
    plt.plot(series.index, series.values, 
             label=wr, 
             color=wr_colors.get(wr, None),
             marker='o', 
             linewidth=3, 
             alpha=0.6)

overall_avg = df_long.groupby('Year')['Value'].mean()
plt.plot(overall_avg.index, overall_avg.values, 
         label='Average', 
         color='grey', 
         linestyle='--', 
         linewidth=1, 
         alpha=1)

plt.title("Market Value Evolution by Work Rate (2019 - 2023)", fontsize=18, fontweight='bold', pad=25)
plt.ylabel("Average Value (€)", fontsize=12, fontweight='bold')
plt.xlabel("Year", fontsize=12, fontweight='bold')

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'€{x*1e-6:,.0f}M'))

plt.legend(title="Work Rate", frameon=False, loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.tight_layout()

plt.show()

```

```python
grouped1 = filtered_df.groupby("Work Rate")["ID"].agg("count")
grouped1.sort_values(ascending=False)
```

```python
custom_order = ['High/ High', 'High/ Medium', 'High/ Low','Medium/ High', 'Medium/ Medium', 'Medium/ Low','Low/ High', 'Low/ Medium', 'Low/ Low']
plot_data = grouped1.reindex(custom_order)[::-1]
#plot_data

wr_colors = {
    'High/ High': '#E65100',    # Strong Orange
    'High/ Medium': '#FF9800',  # Average Orange
    'High/ Low': '#FFCC80',     # Weak Orange
    'Medium/ High': '#1B5E20',  # Strong Green
    'Medium/ Medium': '#4CAF50',# Average Green
    'Medium/ Low': '#A5D6A7',   # Weak Green
    'Low/ High': '#0D47A1',     # Strong Blue
    'Low/ Medium': '#2196F3',   # Average Blue
    'Low/ Low': '#BBDEFB'       # Weak Blue
}
colors = [wr_colors[label] for label in plot_data.index]

plt.figure(figsize=(14, 8))

bars = plt.bar(plot_data.index, plot_data.values, color=colors, edgecolor='#333333', linewidth=1)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, 
             height + 15,            
             f'{int(height)}', 
             ha='center',            
             va='bottom',            
             fontweight='bold', 
             color='#333333')

plt.title("Number of Players by Work Rate", fontsize=20, fontweight='bold', pad=30, color='#1A1A1A')

plt.xlabel("Count of Players", fontsize=13, fontweight='bold', color='#333333')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_color('#CCCCCC')
plt.gca().spines['left'].set_color('#CCCCCC')

plt.tight_layout()

plt.show()
```

```python



identity_columns = ['ID', 'Name', 'Work Rate']
value_columns = ['2019 Value', '2020 Value', '2021 Value', '2022 Value', '2023 Value']

df_long2 = filtered_df.melt(id_vars=identity_columns, value_vars=value_columns, var_name='Year', value_name='Value')
df_long2['Year'] = df_long2['Year'].str.replace(' Value', '', regex=False).astype(int)
df_long2



low_df = df_long2[df_long2['Work Rate'] == 'Low/ Low']
category_total = low_df.groupby('Year')['Value'].sum()

messi_id = 158023 
messi_data = low_df[low_df['ID'] == messi_id].groupby('Year')['Value'].sum()
messi_data

```

```python

plt.figure(figsize=(14, 7))

plt.fill_between(category_total.index, category_total.values, color='#BBDEFB', alpha=0.4, label='Total Low/ Low Market Value')
plt.fill_between(messi_data.index, messi_data.values, color='#1976D2', alpha=0.8, label='Messi (Low/ Low Entry)')

plt.title("Messi's Value vs. Total Low/Low Category", fontsize=18, fontweight='bold', pad=25)
plt.ylabel("Cumulative Value (€)", fontsize=12, fontweight='bold')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'€{x*1e-6:,.1f}M'))

plt.legend(loc='upper right', frameon=False)
plt.grid(axis='y', linestyle=':', alpha=0.5)
plt.grid(axis='x', linestyle='-', alpha=0.2)
plt.tight_layout()
plt.xticks(category_total.index, [str(int(year)) for year in category_total.index])
plt.show()
```
