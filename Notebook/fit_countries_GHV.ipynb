```python
country_area_df = (
pd.DataFrame(list(country_to_area.items()),
columns=["Nationality", "area"])
.sort_values("Nationality")
.reset_index(drop=True)
)
```

```python
all_countries = df["Nationality"].dropna().unique()


missing_countries = set(all_countries) - set(country_area_df["Nationality"])


missing_df = pd.DataFrame({
"Nationality": sorted(missing_countries),
"area": "Other"
})


country_area_df = (
pd.concat([country_area_df, missing_df], ignore_index=True)
.sort_values("Nationality")
.reset_index(drop=True)
)
```
