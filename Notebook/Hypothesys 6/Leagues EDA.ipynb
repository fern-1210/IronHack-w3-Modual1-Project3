```python
df_wanted = df[["ID","Name","Wage","2023 Club"]]
df_wanted
df_wage_league = pd.merge(df_wanted,leagues,how="left",left_on="2023 Club", right_on="Team (DB)")
df_wage_league = df_wage_league.drop(columns=["Team","Match Score","Team (DB)"])
df_wage_league["Country"] = df_wage_league["Country"].str.upper()
df_wage_league["League"] = df_wage_league["League"].str.upper()
df_wage_league["Country - League"] = df_wage_league["Country"] + " - " + df_wage_league["League"].str.upper()
df_wage_league
```

```python
grouped1 = df_wage_league.groupby("Country - League")[["Wage"]].agg(["sum", "mean", "count"])
grouped1.columns = ["Wage (Sum)", "Wage (Mean)", "Number of Players"]
grouped1 = grouped1.sort_values("Wage (Mean)", ascending=False)
grouped1[["Wage (Sum)", "Wage (Mean)"]]=grouped1[["Wage (Sum)", "Wage (Mean)"]].map('€{:,.0f}'.format)
grouped1 = grouped1.loc[grouped1["Number of Players"]>=50]
grouped1 = grouped1.head(10)
grouped1
```

```python
import matplotlib.pyplot as plt

# 1. Prepare data (numeric)
plot_data = df_wage_league.groupby("Country - League")[["Wage"]].agg(["sum", "mean", "count"])
plot_data.columns = ["Wage (Sum)", "Wage (Mean)", "Number of Players"]
plot_data = plot_data.loc[plot_data["Number of Players"] >= 50]
plot_data = plot_data.sort_values("Wage (Mean)", ascending=False).head(10)

# 2. Setup Figure - making it wider (16x10) helps spread out the bubbles
plt.figure(figsize=(16, 10))

# Data variables
y_values = plot_data["Wage (Sum)"] / 1_000_000
bubble_sizes = (plot_data["Wage (Mean)"] / plot_data["Wage (Mean)"].max()) * 2000

# 3. Plot
sc = plt.scatter(
    plot_data["Number of Players"], 
    y_values, 
    s=bubble_sizes,
    c=plot_data["Wage (Mean)"], 
    cmap="RdYlGn_r", 
    alpha=0.6,          # Lower alpha helps see through overlaps
    edgecolors="black",
    linewidth=0.5
)

# 4. Manual Overlap Fix: The "BBox" Label
for i, league in enumerate(plot_data.index):
    # We alternate the label position (some above, some below) to reduce collision
    offset = 15 if i % 2 == 0 else -20 
    
    plt.annotate(
        league, 
        (plot_data.loc[league, "Number of Players"], y_values.loc[league]),
        xytext=(0, offset), 
        textcoords='offset points', 
        ha='center',
        fontsize=8,
        fontweight='bold',
        # This white box prevents the bubble color from making the text unreadable
        bbox=dict(boxstyle='round,pad=0.2', fc='white', alpha=0.7, edgecolor='none')
    )

# 5. Aesthetics
plt.colorbar(sc).set_label('Average Wage (€)')
plt.title("Total Wage vs Players (Top 10 Leagues)", fontsize=18, pad=25)
plt.xlabel("Number of Players", fontsize=12)
plt.ylabel("Total Wage (€ Millions)", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)

# Increase axis limits slightly so labels near the edges aren't cut off
plt.margins(0.1)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

numeric_data = (df_wage_league.groupby(["Country - League", "2023 Club"])["Wage"]
                .mean()
                .sort_values(ascending=False)
                .head(10))

leagues = [index[0] for index in numeric_data.index][::-1]
clubs = [index[1] for index in numeric_data.index][::-1]
wages = numeric_data.values[::-1]

color_map = {
    "SPAIN - LALIGA SANTANDER": "#F9E076",
    "FRANCE - LIGUE 1 UBER EATS": "#72A0C1",
    "ENGLAND - PREMIER LEAGUE": "#E57373",
    "ITALY - SERIE A TIM": "#81C784",
    "GERMANY - BUNDESLIGA": "#546E7A" 
}

bar_colors = [color_map.get(l, "lightgrey") for l in leagues]

plt.figure(figsize=(12, 9))
bars = plt.barh(clubs, wages, color=bar_colors, edgecolor='black', linewidth=1.2)

for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, 
             f' €{width:,.0f}', 
             va='center', fontweight='bold', fontsize=10)

plt.title("Top 10 Clubs by Avg Wage", fontweight='bold', fontsize=16, pad=20)
plt.xlabel("Average Weekly Wage (€)")

from matplotlib.patches import Patch
legend_elements = [Patch(facecolor=color, edgecolor='black', label=league) 
                   for league, color in color_map.items() if league in set(leagues)]
plt.legend(handles=legend_elements, title="Leagues", loc='lower right')

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.tight_layout()
plt.show()
```
