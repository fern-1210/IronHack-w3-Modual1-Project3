```python
import requests
import pandas as pd

url = "https://www.openligadb.de/api/getmatchdata/pl/2024/1.%20Spieltag"
matches = requests.get(url).json()

teams = {}

for m in matches:
    for side in ["Team1", "Team2"]:
        t = m[side]
        teams[t["TeamId"]] = {
            "team_id": t["TeamId"],
            "team_name": t["TeamName"],
            "short_name": t["ShortName"]
        }

df_teams = pd.DataFrame(teams.values())
df_teams
```
